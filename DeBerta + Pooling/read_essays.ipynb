```python
list1[6]='PATRIOTISM AND SPORT.\n'
list1[7]='AN ESSAY ON TWO CITIES.\n'
list1[23]='SPIRITUALISM.\n'
list1[29]='THE "EATANSWILL GAZETTE."\n'
```

```python
corpus= ''.join(lines1[75:])
```

```python
essays=[]
for i in range(len(list1)-2):
    start = list1[i]
    end = list1[i+1]
    ess= corpus[corpus.rfind(start)+len(start):corpus.rfind(end)]
    essays.append(ess)
```

```python
import pickle
with open("newessays", "wb") as fp:   #Pickling
    pickle.dump(essays, fp)
```
